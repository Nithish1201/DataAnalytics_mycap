```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('survey_data.csv')

plt.figure(figsize=(12,8))
plt.bar(data['Name'].head(10),data['Academics'].head(10),color = 'red',alpha = 0.6)
plt.xticks(color = 'white')
plt.yticks(color = 'white',size = 15)
plt.grid(linewidth = 0.5,alpha = 0.2,color = 'grey')
plt.title('Students Academic Report',color = 'white',size = 30,style = 'oblique')
plt.show()
```
